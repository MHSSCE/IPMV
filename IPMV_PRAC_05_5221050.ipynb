```python
img = cv2.imread('Donut.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray",vmin=0,vmax=255)
plt.title("Grayscale Image")
plt.show()
```

```python
#######################Prewitt in X direction ######################
m = img.shape[0]
n = img.shape[1]
#print(m)
#print(n)
```

```python
#Convolution defination
def convolution(mas, im):
  img = im
  mask = mas
  img_new = np.zeros([m, n])
  for i in range(1, m-1):
    for j in range(1, n-1):
      temp = img[i-1, j-1]*mask[0,0] + img[i-1, j]*mask[0,1] + img[i-1, j+1]*mask[0,2] + img[i, j-1]*mask[1,0] + img[i, j]*mask[1,1] + img[i, j+1]*mask[1,2] + img[i+1, j-1]*mask[2,0] + img[i+1, j]*mask[2,1] + img[i+1, j+1]*mask[2,2]
      img_new[i, j] = temp
      img_new = img_new.astype(np.uint8)
  return img_new
```

```python
#using prewittl mask in X axis

mask_X = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])


img_X_prev = convolution(mas = mask_X, im = img)

plt.figure(figsize = figsize)
plt.imshow(img_X_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('prewitt in X axis')
plt.show()
```

```python
#using prewittl mask in Y axis

mask_Y = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])


img_Y_prev = convolution(mas = mask_Y, im = img)

plt.figure(figsize = figsize)
plt.imshow(img_Y_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('prewitt in Y axis')
plt.show()
```

```python
#prewitt X and Y axis combined
mask_total = img_X_prev + img_Y_prev
plt.figure(figsize = figsize)
plt.imshow(img_X_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('prewittl X and Y axis combined')
plt.show()
```

```python
#using sobel mask in X axis

mask_X = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])


img_X_prev = convolution(mas = mask_X, im = img)

plt.figure(figsize = figsize)
plt.imshow(img_X_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('Sobel in X axis')
plt.show()
```

```python
#using sobel mask in Y axis

mask_Y = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])


img_Y_prev = convolution(mas = mask_Y, im = img)

plt.figure(figsize = figsize)
plt.imshow(img_Y_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('Sobel in Y axis')
plt.show()
```

```python
#Sobel X and Y axis combined
mask_total = img_X_prev + img_Y_prev
plt.figure(figsize = figsize)
plt.imshow(img_X_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('Sobel X and Y axis combined')
plt.show()
```

```python
#using Laplacian mask in X axis

mask_X = np.array([[0, -1, 0], [-1, 4, 1], [0, -1, 0]])


img_X_prev = convolution(mas = mask_X, im = img)

plt.figure(figsize = figsize)
plt.imshow(img_X_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('Laplacian in X axis')
plt.show()
```

```python
#using Laplacian mask in Y axis

mask_Y = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])


img_Y_prev = convolution(mas = mask_Y, im = img)

plt.figure(figsize = figsize)
plt.imshow(img_Y_prev, cmap = 'gray', vmin = 0, vmax = 256)
plt.title('Laplacian in Y axis')
plt.show()
```
